```python
def preprocess_base(df : pd.DataFrame) -> pd.DataFrame:
    df = df.copy()
    df['CreatedDate'] = pd.to_datetime(df['CreatedDate'], errors='coerce')
    df['year']    = df['CreatedDate'].dt.year
    df['month']   = df['CreatedDate'].dt.month
    df['weekday'] = df['CreatedDate'].dt.weekday
    df['weekend'] = (df['weekday'] >= 5).astype(int)


    df = pd.get_dummies(df, columns=['PaymentType', 'service'])


    df.fillna(0)
    df.drop_duplicates(inplace=True)
    df.drop("max_number_of_ordered_items", axis = 1)
    df.drop("min_number_of_ordered_items", axis = 1)

    for col in df.select_dtypes(include='bool').columns:
        df[col] = df[col].astype(int)    


    
    return df
```

```python
def add_features(df : pd.DataFrame) -> pd.DataFrame: 
    df = df.copy()
        
    user_counts = df['user_id'].value_counts()
    df['user_order_count'] = df['user_id'].map(user_counts)
    
    df['ratio_current_to_mean'] = df['total_ordered'] / (df['mean_number_of_ordered_items']+1e-6)

    df['is_new_product'] = (df['NmAge'] <= 30).astype(int)
    # --------

    df['days_between'] = df.groupby('user_id')['CreatedDate'] \
                        .diff().dt.days
    df['user_mean_interval'] = df.groupby('user_id')['days_between'].transform('mean')
    df['user_std_interval']  = df.groupby('user_id')['days_between'].transform('std')

    
    df.drop("CreatedDate", axis = 1, errors='ignore', inplace=True)
    return df
```

```python
def align(a, b):
    all_cols = a.columns.union(b.columns)
    return (a.reindex(columns=all_cols, fill_value=0),
            b.reindex(columns=all_cols, fill_value=0))
    
data_train, data_test = align(data_train, data_test)
```

```python
from sklearn.preprocessing import StandardScaler
num_cols = None    
scaler   = None

def scale_df(df: pd.DataFrame, fit: bool) -> pd.DataFrame:
    global num_cols, scaler
    if num_cols is None:
        num_cols = df.select_dtypes(include=['int64','float64']).columns.difference(['target'])
    if fit:
        scaler = StandardScaler().fit(df[num_cols])
    df[num_cols] = scaler.transform(df[num_cols])
    return df
```

```python

train = preprocess_base(data_train)
test  = preprocess_base(data_test)

train = add_features(train)
test  = add_features(test)

train, test = align(train, test)

train = scale_df(train, fit=True)
test  = scale_df(test, fit=False)

data_train = train
data_test  = test

data_train.info()

```

```python
from sklearn.base import BaseEstimator, TransformerMixin

class BasePreprocessor(BaseEstimator, TransformerMixin):
    """Обёртка вокруг вашей функции preprocess_base."""
    def fit(self, X, y=None):
        return self
    def transform(self, X):
        return preprocess_base(X.copy())

class FeatureAdder(BaseEstimator, TransformerMixin):
    """Обёртка вокруг вашей функции add_features."""
    def fit(self, X, y=None):
        return self
    def transform(self, X):
        return add_features(X.copy())

class ScalerTransformer(BaseEstimator, TransformerMixin):
    """Обёртка вокруг вашей функции scale_df."""
    def __init__(self):
        # здесь можете положить StandardScaler, если нужно
        pass
    def fit(self, X, y=None):
        # если scale_df внутри себя обучает scaler, можно здесь ничего не делать
        return self
    def transform(self, X):
        return scale_df(X.copy())
    
class Aligner(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        # запоминаем, какие колонки были в train
        self.train_cols_ = X.columns
        return self
    def transform(self, X):
        # объединяем колонки train и текущего X, заполняем 0
        all_cols = self.train_cols_.union(X.columns)
        return X.reindex(columns=all_cols, fill_value=0)
        
```

```python

from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_recall_curve, classification_report, precision_score, recall_score, f1_score
from catboost import CatBoostClassifier, Pool 
from sklearn.impute import SimpleImputer 
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LogisticRegression
import optuna, numpy as np
from sklearn.metrics import accuracy_score
from sklearn.model_selection import StratifiedKFold


import matplotlib.pyplot as plt
from catboost import CatBoostClassifier, Pool
from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_score, recall_score

X_tr, y_tr = data_train.drop("target", axis=1), data_train["target"]
X_te, y_te = data_test.drop("target", axis=1),  data_test["target"]

X_trn, X_val, y_trn, y_val = train_test_split(
    X_tr, y_tr,
    test_size=0.2,
    stratify=y_tr,
    random_state=42
)

# 3) Создаём Pool'ы без cat_features
train_pool = Pool(X_trn, y_trn)
val_pool   = Pool(X_val, y_val)
test_pool  = Pool(X_te,  y_te)

def objective(trial):
    params = {
        'iterations': trial.suggest_int('iterations', 500, 2000),
        'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.3, log=True),
        'depth': trial.suggest_int('depth', 2, 3),
        'l2_leaf_reg': trial.suggest_float('l2_leaf_reg', 10**7, 10**7+2),
        'border_count': trial.suggest_categorical('border_count', [32, 64, 128, 254]),
        'bagging_temperature': trial.suggest_float('bagging_temperature', 0, 1),
        'subsample': trial.suggest_float('subsample', 0.6, 1.0),
        'random_strength': trial.suggest_float('random_strength', 1e-9, 10, log=True),
        'loss_function': 'Logloss',
        'eval_metric': 'F1',
        'random_seed': 42,
        'verbose': False,
        'task_type': 'CPU'
    }

    model = CatBoostClassifier(
        **params
    )
    model.fit(
        X_trn, 
        y_trn, 
        eval_set = (X_val, y_val), 
        early_stopping_rounds=100,
        verbose=False
    )
    preds = model.predict(X_val)
    return f1_score(y_val, preds)
    '''
    cv_scores = []
    cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
    
    for train_idx, test_idx in cv.split(X_tr, y_tr):
        train_pool = Pool(X_tr.iloc[train_idx], y_tr.iloc[train_idx])
        val_pool = Pool(X_tr.iloc[test_idx], y_tr.iloc[test_idx])
        
        model.fit(train_pool, eval_set=val_pool)
        preds = model.predict(val_pool)
        cv_scores.append(f1_score(y_tr.iloc[test_idx], preds))
    return np.mean(cv_scores)
    '''
    
    
    
```

```python
study = optuna.create_study(
    direction='maximize',
    sampler=optuna.samplers.TPESampler(seed=42),  # Используем интеллектуальный сэмплер
    pruner=optuna.pruners.MedianPruner(n_warmup_steps=10)
)
```

```python
study.optimize(
    objective,
    n_trials=10,  # Количество итераций оптимизации
    timeout=3300,  # Максимальное время выполнения в секундах (1 час)
    show_progress_bar=True  # Визуализация прогресса
)
```

```python
# 4. Выводим результаты лучшего trial
print("\nBest trial:")
trial = study.best_trial
print(f"  Mean F1 (CV): {trial.value:.4f}")
print("  Best params:")
for key, value in trial.params.items():
    print(f"    {key}: {value}")

# 5. (Опционально) Визуализация результатов
#optuna.visualization.plot_optimization_history(study).show()
#optuna.visualization.plot_param_importances(study).show()
data_train.info()
```

```python
from sklearn.metrics import roc_curve, roc_auc_score, auc

final_params = study.best_trial.params.copy()


'''final_params.update({
    'loss_function': 'Logloss',
    'eval_metric': 'F1',
    'random_seed': 42,
    'verbose': False
})'''

final_model = CatBoostClassifier(**final_params)
final_model.fit(Pool(X_tr, y_tr))

# Полная оценка на тесте
test_preds = final_model.predict(test_pool)

# 1) Получаем вероятности «1»-го класса на тесте
probs = final_model.predict_proba(X_te)[:, 1]

# 2) Строим precision–recall кривую
precision, recall, thresholds = precision_recall_curve(y_te, probs)

# 3) Выбираем пороги, где recall >= 0.1
mask = recall[:-1] >= 0.1  # thresholds имеет длину len(precision)-1

# 4) Среди них находим индекс максимального precision
best_idx = mask.nonzero()[0][np.argmax(precision[:-1][mask])]
best_threshold = thresholds[best_idx]
best_precision = precision[:-1][mask][np.argmax(precision[:-1][mask])]
best_recall    = recall[:-1][mask][np.argmax(precision[:-1][mask])]

print(f"\nOptimal threshold={best_threshold:.4f} → precision={best_precision:.3f}, recall={best_recall:.3f}")

# 5) Делаем финальное предсказание с новым порогом
adjusted_preds = (probs >= best_threshold).astype(int)

# 6) Смотрим итоговый отчёт
print("\nFull Test Report with adjusted threshold:")
print(classification_report(y_te, adjusted_preds))
```
